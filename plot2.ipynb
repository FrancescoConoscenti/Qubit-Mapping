```python
import matplotlib.pyplot as plt
import numpy as np

# Define explicit x and y values for the three datasets
x_values =  [3,       5,        7,      9,      15]
original =  [7,       11,       10,     12,     15]
lin_smt =   [0,       32,       15,     28      ,0]
heuristic = [18,      39,       34,     56,     181]
ilp =       [11,      17,       13,     26,      39]
sabre =     [13,      27,       14,     34,      39]

# Define the width of the bars
bar_width = 0.14

# Create an array of x positions for the first set of bars
r1 = np.arange(len(x_values))

# Create arrays of x positions for the second and third sets of bars, offset by the bar width
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]
r5= [x + bar_width for x in r4]

# Create the bar plot
plt.bar(r1, original, color='grey', width=bar_width,  label='Original')
plt.bar(r2, lin_smt, color='green', width=bar_width,  label='Lin SMT')
plt.bar(r3, ilp, color='blue', width=bar_width,  label='ILP')
plt.bar(r4, heuristic, color='orange', width=bar_width,  label='Heuristic')
plt.bar(r5, sabre, color='purple', width=bar_width,  label='SABRE')

# Add labels and title
plt.xlabel('Number of Qubits', )
plt.ylabel('Depth')
plt.title('Depth of mapped circuit')
plt.xticks([r + bar_width for r in range(len(x_values))], x_values)

# Add legend
plt.legend()

# Show the plot
plt.show()
```
